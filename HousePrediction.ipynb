```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#loading the csv file
df =pd.read_csv('homeprices.csv')
```

```python
df
```

```python
#plotting a scatter plot to visualise the data
plt.xlabel('Area')
plt.ylabel('Prices')
plt.scatter(df['area'],df['price'],color='red',marker='+')
```

```python
#building a regression model
reg = linear_model.LinearRegression()

#making sure area is used as a 2d array
reg.fit(df[['area']],df.price)
```

```python
#Predicting the house price for an area of 3300
reg.predict([[3300]])
```

```python
#finding the slope
reg.coef_
```

```python
#finding the y intercept
reg.intercept_
```

```python
#loading the areas.csv file
new_df = pd.read_csv('areas.csv')
new_df
```

```python
#assigning a variable p to the predicted values for the areas.csv file
p =reg.predict(new_df)
```

```python
new_df['prices']= p
new_df
```

```python
#converting the predictions made for areas.csv file into a csv file
new_df.to_csv('Prediction.csv',index= False)
```
